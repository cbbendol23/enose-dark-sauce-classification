```python
import pandas as pd
import numpy as np
import joblib

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
# ------------------------
# STEP 2: Load validation dataset
# ------------------------
val_df = pd.read_csv("validationset_article2.csv")

# Same target column name as training
X_val = val_df.drop(columns=["Label"])
y_val = val_df["Label"]

print("Validation set shape:", val_df.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)

# ------------------------
# STEP 3: Load saved ensemble model + label encoder
# ------------------------
ensemble_model = joblib.load("ensemble_model.joblib")
le = joblib.load("label_encoder.joblib")   # this should be the same LabelEncoder you used in training

# ------------------------
# STEP 4: Encode validation labels (because model predicts encoded classes)
# ------------------------
y_val_enc = le.transform(y_val)

# ------------------------
# STEP 5: Predict (encoded) then decode for reporting
# ------------------------
y_pred_enc = ensemble_model.predict(X_val)
y_pred = le.inverse_transform(y_pred_enc)

# ------------------------
# STEP 6: Evaluate using decoded (string) labels like your training code
# ------------------------
print("\nEnsemble Validation Accuracy:", accuracy_score(y_val, y_pred))
print("\nClassification Report (Validation):")
print(classification_report(y_val, y_pred))

# Confusion Matrix (label strings, consistent label order)
cm = confusion_matrix(y_val, y_pred, labels=le.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(xticks_rotation=45)
plt.title("Confusion Matrix - Ensemble (Validation Set)")
plt.tight_layout()
plt.show()
```
